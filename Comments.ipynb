#**Comments**
* Comments are used to increase the readability of the code.
* These are completely ignored by the python interpreter.
* They do not display or affect the output of the code.
* Its always a good practice to add coments in your code.
* Different types of comments are discussed below with examples.

***Single Line Comments:***  
* Starts with a hash symbol'#'followed by the text to be written as a comment that lasts till the end of the line.

```python
#calculating average
num1=10
num2=20
(num1+num2)/2
```

***Multiline Comments***
 * While writing multiline comments ,we begin the comment with three single quotes''' or three double quotes""" followed bythe text spread across multiple lines.the comments ends with the same three single or double quotes''' or """.

```python
"""This program calculates
the average of two numbers stored in\
two different variables"""
num1=10
num2=20
(num1+num2)/2
```
